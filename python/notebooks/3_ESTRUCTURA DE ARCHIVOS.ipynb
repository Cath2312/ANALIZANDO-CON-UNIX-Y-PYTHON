## Estructura de Archivos
Una vez extra√≠dos, podemos visualizar la estructura del directorio:

```python
import os

def print_directory_structure(root_dir, level=0):
    for item in os.listdir(root_dir):
        path = os.path.join(root_dir, item)
        print("  " * level + "- " + item)
        if os.path.isdir(path):
            print_directory_structure(path, level + 1)

print_directory_structure('temp_aclImdb')
```
